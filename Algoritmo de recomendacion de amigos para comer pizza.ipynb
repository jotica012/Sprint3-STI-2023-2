```python
import pandas as pd
data = pd.read_csv('basepizza.csv')
data.head()
```

```python
from sklearn.neighbors import NearestNeighbors
k = 5 
features = data[['Pollo', 'Tocineta', 'Jamón', 'Cábano', 'Chorizo', 'Calamar', 'Pepperoni', 'Piña', 'Tomate', 'Maicitos', 'Plátano maduro', 'Jalapeño', 'Cebolla', 'Champiñón', 'Aceitunas']]
knn = NearestNeighbors(n_neighbors=k)
knn.fit(features)
usuario_ejemplo = 0
distancias, indices = knn.kneighbors([features.iloc[usuario_ejemplo]])
print("Usuarios más cercanos:")
print(data.iloc[indices[0]])
print("Distancias:")
print(distancias[0])
```
